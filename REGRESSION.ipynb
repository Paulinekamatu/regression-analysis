```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import datasets,linear_model
import seaborn as sns
```

```python
clientspending=pd.read_csv("clientspending.csv")
print(clientspending)
```

```python
trainingset,testingset=train_test_split(clientspending,test_size=0.3,random_state=2)
print('training data is:\n',trainingset)

#test_sie=0.3 specifies that 30% of the data should be used as testing set, 70% for training set.
#random_state=2 ensures the setting of the random number generator


print('testing data is:\n',testingset)
```

```python
trainingindependentvariable=trainingset.values[:,0:1]
trainingdependentvariable=trainingset.values[:,0:1]
print('independent variables for the training data:\n',trainingindependentvariable)
print('dependent variables for the training data:\n',trainingdependentvariable)
```

```python
testingindependentvariable=trainingset.values[:,0:1]
testingdependentvariable=trainingset.values[:,0:1]
print('independent variables for the testing data:\n',trainingindependentvariable)
print('dependent variables for the testing data:\n',trainingdependentvariable)
```

```python
plt.scatter(trainingindependentvariable,trainingdependentvariable,color='blue')
plt.title('client data for training data set')
plt.xlabel('annual income')
plt.ylabel('expenditure')
plt.show()
```

```python
plt.scatter(testingindependentvariable,testingdependentvariable,color='red')
plt.title('client data for testing')
plt.xlabel('annual income')
plt.ylabel('expenditure')
plt.show()
```

```python
regression=linear_model.LinearRegression()
regression.fit(trainingindependentvariable,trainingdependentvariable)
print('coefficient of regression:\n',regression.coef_)
sns.regplot(x=trainingindependentvariable,y=trainingdependentvariable,data=clientspending)
plt.show()
```
